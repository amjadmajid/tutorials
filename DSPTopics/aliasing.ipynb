```python
# File name: aliasing.py
# Author: Amjad Yousef Majid
# email: com_amjed@live.nl
# Date created: 12/25/2015
# Python Version: 3.5

import matplotlib.pyplot as plt # For ploting
import numpy as np # to work with numerical data efficiently
import math as m

def signal(f, fs):
    w = 2*np.pi*f
    return [ np.sin(w * (i/fs) ) for i in np.arange(fs)] 

fs = 20 # sample rate 
f = 4 
x = np.arange(fs) # the points on the x axis for plotting
s_f_4 = signal(f, fs)

s_f_4_h = signal(f, fs*10) # higher sampling to generate higher resolution waveform
x_h = np.arange(fs * 10) # this step to show the waveform 

f = 16 
s_f_16 = signal(f, fs)
s_f_16_h = signal(f, fs*10) # higher sampling to generate higher resolution waveform

 
f = 24 
s_f_24 = signal(f, fs)
s_f_24_h = signal(f, fs*10) # higher sampling to generate higher resolution waveform

 

% matplotlib inline
plt.figure(1)

plt.subplot(321)
plt.plot(x_h,s_f_4_h,'y')
plt.subplot(322)
plt.plot(x,s_f_4)


plt.subplot(323)
plt.plot(x_h,s_f_16_h, 'y')
plt.subplot(324)
plt.plot(x,s_f_16)

plt.subplot(325)
plt.plot(x_h,s_f_24_h, 'y')
plt.subplot(326)
plt.plot(x,s_f_24)
```
