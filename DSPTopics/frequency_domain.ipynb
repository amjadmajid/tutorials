```python
import matplotlib.pyplot as plt
import math
import numpy as np
import random
%matplotlib inline

plt.rcParams['axes.labelsize']=16

basicFrq = 3
numFrq   = 12
stepFrq  = 1

twoPi = np.pi*2
x=np.linspace(0,twoPi * basicFrq , 100 )

y=[]          # time amplitude
frqAmp=[]     # frequency amplitudes
frq=[]        # frequency components 

for i in range(0,numFrq,stepFrq):
    f = basicFrq + i # * 10   * random.random()
    fAmp = (1.0/f)
    y.append( np.sin( np.array(x * f ) ) * fAmp ) 
    frqAmp.append(fAmp)
    frq.append(f)

#time domain plotting
frqs = []
v = np.array([0] * len(y[0]) )
for j in (y):
    plt.figure(figsize=(10,3) )
    plt.xlabel('Time [sec]')
    plt.ylabel('Amplitude')
    frqs.append(j)
    for kk in frqs:
        plt.plot(x/(np.pi * 2),kk)
        pass
        
    v = v + j
    plt.plot(x/twoPi ,v, 'k', linewidth=3 )
    
# frequency domain plotting
plt.figure()
plt.stem(frq, frqAmp )
plt.xlabel('Hz')
plt.ylabel('Amplitude')
plt.gca().set_xlim([0, frq[-1]+1 ]) 
plt.show()
```
