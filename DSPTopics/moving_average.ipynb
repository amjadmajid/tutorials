```python
"""
Using moving average to smooth out a signal 
@ Amajd Yousef Majid
20/12/2015
"""
import matplotlib.pyplot as plt # For ploting
import numpy as np # to work with numerical data efficiently
import math as m
fs = 100 # sample rate 
f = 2 # the frequency of the signal

x = np.arange(fs) # the points on the x axis for plotting
# compute the value (amplitude) of the sin wave at the for each sample
y = [ ( np.random.rand() + np.sin(2*np.pi*f * (i/fs)) ) 
     for i in np.arange(fs)]
j = 0
ave = []
for i in range(len(y)-4): 
    j+=1
    ave.append( (y[j-1]+y[j] + y[j+1] + y[j+2] + y[j+3] )/5 ) 
ave.append( ave[-1] )
ave.append( ave[-2] )
ave.append( ave[-3] )
ave.append( ave[-4] )

% matplotlib inline
# showing the exact location of the smaples
#plt.stem(x,y, 'r', )
plt.plot(x,y)
plt.plot(x,ave,'r')
```
