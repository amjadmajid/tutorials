```python
# draw a line
import matplotlib.pyplot as plt
% matplotlib inline
#                               Point 1  Point 2
plt.gca().add_line(plt.Line2D( (  0,      0.5), 
                               (  1,      0.5) ))

plt.grid(True)
plt.show()
```

```python
# These are all straight lines 
import matplotlib.pyplot as plt
% matplotlib inline

N = 16
for i in range(N):
    plt.gca().add_line(plt.Line2D( (0,i), (N-i, 0), color='0.75' ) )

plt.grid(True)
plt.axis('scaled')    # generate equal axes and are not tight (the default config)
plt.show()
```

```python
#drawing primitive shapes
import matplotlib.patches as patches 
import matplotlib.pyplot as plt

#Circle
shape = patches.Circle((0,0), radius=1.0, color='0.75')
plt.gca().add_patch(shape)

#Rectangle
shape = patches.Rectangle((1,1),2.0,1.0, color='k', fill=False)
plt.gca().add_patch(shape)

#Fancy box
shape = patches.FancyBboxPatch((-3,-3),2.0,1.0, color='0.75', 
                               boxstyle='sawtooth')  # larrow, rarrow, round, round4, roundtooth, square
plt.gca().add_patch(shape)

#Ellipse
shape = patches.Ellipse((-3,1), 2.0,1.0, angle=45. , color='0.75')
plt.gca().add_patch(shape)


plt.axis('scaled')
plt.show()

```

```python
#drawing Polygon
import matplotlib.patches as patches 
import matplotlib.pyplot as plt

#Polygon
shape = patches.Polygon([(0,0), (0,1), (1,1),(0.5,0.5), (1,0) ,(0,0)],
                        fill=None,
                        lw=8,          
                        ls='dashed',
                        edgecolor='0.75')
plt.gca().add_patch(shape)


plt.axis('scaled')
plt.show()
```
