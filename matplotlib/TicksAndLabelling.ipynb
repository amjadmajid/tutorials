```python
# statically assign strings to ticks
import numpy as np
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt

%matplotlib inline

names     = ('America', 'Russia', 'Germany', 'Canada', 'Iraq', 'Morocco')
rand_stat = np.random.randint(0,99, size = len(names))  # random data
pos_list  = np.arange(len(names))  # names positions

# Approach A
ax = plt.axes()  # get the axes handler
ax.xaxis.set_major_locator(ticker.FixedLocator(pos_list)) # set the locations of the ticks
ax.xaxis.set_major_formatter(ticker.FixedFormatter(names)) # set the labels names
# end of Approach A

# Approach B
# plt.xticks(pos_list, names)
# end 

plt.bar(pos_list, rand_stat, color='0.75', align='center')
plt.show()
```

```python
# Using FuncFormatter to dynamically assign strings to 
# the ticks the matplotlib generates

import numpy as np
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt

%matplotlib inline


# defube the function that forms the strings
def maker_label(value, pos):  # the function has to accept two arguments 
      return"{:0.1f}%".format(100. * value)
    
ax = plt.axes()
ax.xaxis.set_major_formatter(ticker.FuncFormatter(maker_label))

data = np.linspace(0,1,100)

plt.plot(data,np.exp(-10 * data), c='k')

# rotate the labels
labels = ax.get_xticklabels()  # get handle to the x axis labels
plt.setp(labels, rotation=30.) # rotate the labels

plt.show()
```
